```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

import plotly.express as px
import plotly.graph_objects as go
from jupyter_dash import JupyterDash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
```

```python
#Read the csv file of data
df = pd.read_csv('Clean_TurkishTV_Data.csv')

#Show the first five rows in the dataset
df.head()
```

```python
import dash
import dash_bootstrap_components as dbc
#from dash.dependencies import Input, Output
from dash import Input, Output, dcc, html

app = dash.Dash(external_stylesheets=[dbc.themes.BOOTSTRAP])

app.title = "Turkish TV Series"
server = app.server

# the style arguments for the sidebar. We use position:fixed and a fixed width
SIDEBAR_STYLE = {
    "position": "fixed",
    "top": 0,
    "left": 0,
    "bottom": 0,
    "width": "16rem",
    "padding": "2rem 1rem",
    "background-color": "#f8f9fa",
}

navbar = dbc.NavbarSimple(
       children=[
        dbc.NavItem(dbc.NavLink("Networks", href="/", active="exact")),
        dbc.NavItem(dbc.NavLink("Genre",
                    href="/page-1", active="exact")),
    ],
    brand="Turkish TV Series",
    brand_href="#",
    color="#000000",
    dark=True,
)
# the styles for the main content position it to the right of the sidebar and
# add some padding.
CONTENT_STYLE = {
    "margin-left": "2rem",
    "margin-right": "2rem",
    "padding": "2rem 1rem",
}


# The controls of Salary page
controls_page0 = dbc.Card(
     [

        html.Label('Adaptation'),
        dbc.Card([dbc.RadioItems(
            id='adaptation_radio_items',
            value=1,
            options=[{
                'label': 'Yes',
                'value': 1
            },
                {
                    'label': 'No',
                    'value': 0
            },

            ],

        )]),


        html.Br(),

        html.Label('Slider for the years'),
                       dcc.Slider(
                                id = "slide",
                                min=1993,
                                max=2021,
                                step=None,
                                marks={
                                     1993:'1993',
                                    
                                     1996:'1996',
                                   
                                     1999:'1999',
                                     
                                     2002: '2002',
                                     
                                     2005: '2005',
                                     
                                     2008: '2008',
                                     
                                     2011: '2011',
                                    
                                     2014: '2014',
                                    
                                     2017: '2017',
                                    
                                     2021: '2021',
                                     },  
                                value= 2017,
                             ), 
                       html.Br([]),
    ],
    body=True,
)

# The controls of Primary Database page
controls_page2 = dbc.Card(
   
)

# The controls of Career Plans page
controls_page3 = dbc.Card(
    
)

content = html.Div(id="page-content", style=CONTENT_STYLE)

app.layout = html.Div([dcc.Location(id="url"), navbar, content])


@app.callback(Output("page-content", "children"), [Input("url", "pathname")])
def render_page_content(pathname):
    
    if pathname == "/":
        return [
            dbc.Row(
                [
                    dbc.Col(
                        width=4,
                        children=dbc.Card(
                            [dbc.CardHeader("Controls"),
                             dbc.CardBody(controls_page0), ]
                        ),
                    ),
                    dbc.Col(
                         width=8,
                            children=dbc.Card(
                                [
                                    dbc.CardHeader("The Visualization"),
                                    dbc.CardBody(dcc.Graph(id="networks"),
                                                 style={"height": "100%"}),
                                ],
                                style={"height": "80vh"},
                            ),
                        
                    ),
                ],
            ),

        ]
    elif pathname == "/page-1":
        return [
             dbc.Row(
                [
                    dbc.Col(
                        width=4,
                        children=dbc.Card(
                            [dbc.CardHeader("Controls"),
                             dbc.CardBody(controls_page0), ]
                        ),
                    ),
                    dbc.Col(
                         width=8,
                            children=dbc.Card(
                                [
                                    dbc.CardHeader("The Visualization"),
                                    dbc.CardBody(dcc.Graph(id="genre"),
                                                 style={"height": "100%"}),
                                ],
                                style={"height": "80vh"},
                            ),
                        
                    ),
                ],
            ),

        ]

       

      

    

    # If the user tries to reach a different page, return a 404 message
    return dbc.Jumbotron(
        [
            html.H1("404: Not found", className="text-danger"),
            html.Hr(),
            html.P(f"The pathname {pathname} was not recognised..."),
        ]
    )



@app.callback(
    Output('networks', 'figure'),

    [Input("adaptation_radio_items", "value"),Input("slide", "value"),]
)

def update_figure(adaptation,years):
    data = df.copy()

    data = data[data['adaptation'] == adaptation]
    data = data[data['year'] <= years]

    fig= px.bar(
        data,
         x= "networks",
         labels={"networks": "Network Name", "count": "Numbers of Series"},
         color='year',
    )
    fig.update_traces(
                  marker_line_width = 0)
    
    fig.update_layout(
        title={
            'text': "Networks That Prsented These Series TV",
            'y': 0.95,
            'x': 0.5,
            'xanchor': 'center',
            'yanchor': 'top'}) 



    
    return fig

@app.callback(
    Output('genre', 'figure'),

    [Input("adaptation_radio_items", "value"),Input("slide", "value"),]
)

def update_figure(adaptation,years):
    data = df.copy()

    data = data[data['adaptation'] == adaptation]
    data = data[data['year'] <= years]

    fig= px.bar(
        data,
         x= "genre1",
         labels={"genre1": "Genre", "count": "Numbers of Series"},
         color='year',
    )
    fig.update_traces(
                  marker_line_width = 0)
    
    fig.update_layout(
        title={
            'text': "Genre of These Series TV",
            'y': 0.95,
            'x': 0.5,
            'xanchor': 'center',
            'yanchor': 'top'}) 



    
    return fig




if __name__ == "__main__":
    app.run_server(port=8888)
```
